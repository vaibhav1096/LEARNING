```python
#answer1
a="12 ** 12"
def arithmetic_operation(a):
    lis=a.split()
    a=int(lis[0])
    b=int(lis[2])
    dcop={'+':a+b,'-':a-b,'*':a*b,'/':a/b,'%':a%b,'**':a**b}
    return dcop[lis[1]]
arithmetic_operation(a)
```

```python
#answer2 
import math
def distance(a,b):
    return math.sqrt((a[0]-b[0])**2 + (a[1]-b[1])**2)

def perimeter(l):
    a=l[0]
    b=l[1]
    c=l[2]
    return round(distance(a,b)+distance(c,b)+distance(a,c),2)

perimeter([[-10, -10], [10, 10 ], [-10, 10] ] )
```

```python
#answer3
a= [
  [0, 1, 0, 0],
  [0, 1, 0, 0],
  [0, 1, 1, 0],
  [1, 1, 1, 1]

]

def tower_height(a):
    n=len(a)
    lis=[]
    for i in range(n):
        sum=0
        for j in range(n):
            sum = sum + a[j][i]
        lis.append(sum)
    return max(lis)
      
tower_height(a)
```

```python
#answer4
def bonus(days): 
    if days > 0 and days <=32:
        sum=0
    elif days>32 and days <=40:
        sum=32*0+(days-32)*325
    elif days>40 and days <=48:
        sum=32*0+8*325+(days-40)*550
    elif days >48:
        sum=32*0+8*325+8*550+(days-48)*600
    return sum
```

```python
bonus(50)
```

```python
#answer5
def is_disarium(n):
    l=str(n)
    k=list(map(int,l))
    sum=0
    for i,j in enumerate(k):
        sum=sum+(j**(i+1))
    if sum==n:
        return True
    else:
        return False
```

```python
is_disarium(135)
```
