```python

# Make a list of each Boolean operator's truth tables 
# (i.e. every possible combination of Boolean values for the operator and what it evaluate ).

l=[True,False]
k=[True,False]
def andor(l,k):
    m=[]
    x=[]
    for i in l:
        for j in k:
            print(l[i],k[i])
            m.append(l[i] and k[j])
            x.append(l[i] or k[j])
    return m,x
    
```

### ANSWER 9

Create a programme that prints. If 1 is stored in spam, prints Hello; if 2 is stored in spam, prints Howdy; and if 3 is stored in spam, prints Salutations! if there's something else in spam.

```python

def spamcheck(spam):
    if spam==1:
        return "Hello"
    elif spam==2:
        return 'Howdy'
    elif spam ==3:
        return 'Salutations'
    else:
        return 'Anything is Something'
        
for i in range(1,5):
    print(i, spamcheck(i))
```
