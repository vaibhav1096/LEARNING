```python
# Answer 3
def censor_string(s,l,c):
    list_of_string=s.split()
    for i in l:
        if i in list_of_string:
            a=list_of_string.index(i)
            list_of_string[a]=c*len(i)
    return ' '.join(list_of_string)
    
```

```python
# Answer 4
def is_polydivisible(n):
    a=str(n)
    l=[]
    if a[0]=='0':
        return False

    for i in range(len(a)):
        if int(a[0:(i+1)])%(i+1)==0:
            l.append(True)
        else:
            l.append(False)
        
    if sum(l)==len(l):
        return True
    else:
        return False
        
```

```python
#Answer 5
def checkprime(n):
    """
    check till n/2 if a factor exists between 0 to n/2 then there must be a factor between n/2 to n
    """
    if n==2 or n==3:
        return True
    a=int(n/2)+1
    for i in range(2,a):

        if n%i==0:
            return False
        elif i==a-1:
            return True
def sum_primes(l):
    sum=0
    for i in l:
        if checkprime(i)==True:
            sum=sum+i
    if sum==0:
        return 'None'
    else:
        return sum
```
