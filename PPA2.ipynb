```python
#answer1
def pentagon(n):
    dots=1
    for i in range(0,n):
        dots=dots+5*i
    return dots
pentagon(8)
```

```python
#answer2
def encrypt(n):
    dicn={'a':0,'e':1,'i':2,'o':3,'u':4}
    n=n[::-1]
    op=""
    for i in n:
        if i not in dicn:
            op=op+i
        else:
            op=op+str(dicn[i])
    op=op+'aca'
    return op
encrypt('banana')
```

```python
#answer3
import datetime
def has_friday_13(month,year):
    x=datetime.datetime(year,month,13)
    if x.strftime("%A")=='Friday':
        return True
    else:
        return False
    
```

```python
#answer5
def pluralize(lis):
    a=set()
    for i in lis:
        if lis.count(i) > 1:
            a.add(i+'s')
        else:
            a.add(i)
    return a
    
pluralize(['arm','aa','a','arm'])
```
